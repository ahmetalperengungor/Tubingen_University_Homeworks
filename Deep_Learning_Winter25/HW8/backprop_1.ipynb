We start by importing the required packages.

````python
import numpy as np
````

We define the input ```X``` and output ```y_real```.

````python
X = np.array([[0, 1], [1, 1], [1, 0], [0, 0]], dtype=float)
y_real = np.array([[0, 1], [1, 0], [0, 1], [1, 0]])
````

Now we define the required Python functions we need later. Most of these function you already saw in the assignment #4.

````python
def tanh(x):
    #TODO: Define the tanh activation function
    return np.tanh(x)

def softmax_crossentropy(a, y: np.ndarray):
    an = (a.T - np.amax(a, axis=-1)).T
    return np.log(np.sum(np.exp(an), axis=-1)) - an[np.where(y == 1)]

def softmax(a):
    an = (a.T - np.amax(a, axis=-1)).T
    expa = np.exp(an)
    return (expa.T/np.sum(expa, axis=-1)).T

def grad_softmax_crossentropy(a, y):
    return softmax(a) - y
````

Next we define the learning parameters.

````python
learning_rate = 1
epochs = 1000
````

Since we are using pure Python, we have to set up the network's weights and biases.

````python
#TODO: Initialise the weight matrices and bias vectors

# W1 should have shape (2, 3)
W1 = np.array(np.random.normal(size=(2,3)), dtype=float)

# b1 should have shape (1, 3)
b1 = np.zeros((1,3), dtype=float)

# W2 should have shape (3, 2)
W2 = np.array(np.random.normal(size=(3,2)), dtype=float)

# b2 should have shape (1, 2)
b2 = np.zeros((1,2), dtype=float)
````

Now, we define the forward pass and the evaluation.

````python
def forward(x):
    u1 = x @ W1 + b1
    z = tanh(u1)
    u2 = z @ W2 + b2
    y = softmax(u2)
    return y    

def evaluate(x, y_real):
    u1 = x @ W1 + b1
    z = tanh(u1)
    u2 = z @ W2 + b2
    loss = softmax_crossentropy(u2, y_real)
    return loss
````

Finally the learning and your work comes into play.

````python
print(50*"+")
print("Before training:")
print("Prediction:", forward(X))
print(50*"+")

for epoch in range(epochs):
    for sample_number in range(X.shape[0]):
        X_batch = np.array([X[sample_number]], dtype=float)
        y_batch = np.array([y_real[sample_number]])

        # Calculate forward pass
        u1 = X_batch @ W1 + b1
        z = tanh(u1)
        u2 = z @ W2 + b2
        loss = softmax_crossentropy(u2,  y_batch)

        #TODO: implement the remaining backpropagation steps here

        # Gradients for backward pass
        dLdu2 = grad_softmax_crossentropy(u2, y_batch)
        delta2 = dLdu2
        dLdW2 = z.T @ delta2
        dLdb2 = delta2
        delta1 = dLdu2 @ W2.T * (1 - z**2)
        dLdW1 = X_batch.T @ delta1
        dLdb1 = delta1

        # Calculate backward pass
        assert(W1.shape == dLdW1.shape)
        assert(b1.shape == dLdb1.shape)
        assert(W2.shape == dLdW2.shape)
        assert(b2.shape == dLdb2.shape)
        W1 -= learning_rate * dLdW1
        b1 -= learning_rate * dLdb1
        W2 -= learning_rate * dLdW2
        b2 -= learning_rate * dLdb2

        # print("Updated weights and biases:")
        # print("W1:", W1)
        # print("b1:", b1)
        # print("W2:", W2)
        # print("b2:", b2)

        print("While training:")
        print(f"Prediction for sample {sample_number}:", forward(X_batch))
        print(f"Cross entropy loss for sample {sample_number}: ", evaluate(X_batch, y_batch))

print(50*"+")
print("After training:")
print("Final prediction: ", forward(X))
print(50*"+")
print("Final prediction: ", forward(X).round(0))
print(50*"+")
````
